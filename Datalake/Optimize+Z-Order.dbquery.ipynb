```sql
-- See how many files your table current has
DESCRIBE DETAIL lakehouse_demo.silver.customers;

-- Co-locate rows by customer_id and country in the same files
OPTIMIZE silver.customers ZORDER BY (customer_id, country);

-- Good candidates for Z-ORDER: columns you filter on frequently
OPTIMIZE silver.customers ZORDER BY (customer_id);           -- point lookups
OPTIMIZE silver.customers ZORDER BY (country, segment);      -- range + group-by
OPTIMIZE silver.customers ZORDER BY (valid_from);            -- time range queries

-- Bad candidates
OPTIMIZE silver.customers ZORDER BY (is_current);   -- ❌ boolean, only 2 values
                                                     --    terrible cardinality
OPTIMIZE silver.customers ZORDER BY (full_name);    -- ❌ unbounded strings, rarely filtered
```
